```python
#importing Libraries

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the datasets

fake = pd.read_csv('/Users/utsav/Desktop/dlpro/News _dataset/data/Fake.csv')
true = pd.read_csv('/Users/utsav/Desktop/dlpro/News _dataset/data/True.csv')

fake["label"] = 0
true["label"] = 1
```

```python
#Combining the datasets
df = pd.concat([fake, true], ignore_index=True)
df.to_csv("/Users/utsav/Desktop/dlpro/News _dataset/data/news_dataset.csv", index=False)
```

```python
#Distribution of labels plot
sns.countplot(x='label', data=df)
plt.title('fake vs real news')
plt.show()
```

```python
#Distribution of Subjects
plt.figure(figsize=(12,6))
sns.countplot(x='subject', data=df)
plt.title('Distribution of Subjects')
plt.show()
```

```python
#Adding length of text and title columns   
df["text_length"] = df["text"].apply(lambda x: len(str(x).split()))
df["title_length"] = df["title"].apply(lambda x: len(str(x).split()))
```

```python
#Text Length Distribution Plot
plt.figure(figsize=(10,5))
sns.histplot(data=df, x="text_length", bins=50, hue="label", kde=True)
plt.title("Text Length Distribution (Fake vs Real)")
plt.show()
```

```python
#Title Length Distribution Plot
plt.figure(figsize=(6,5))
sns.heatmap(df[["label", "text_length", "title_length"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
